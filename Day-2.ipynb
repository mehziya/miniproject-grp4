```python
f= open('marks.txt','r')
f2=open('total.txt','w')
s=f.readlines()
print(s)
for i in s:
    words=i.split()
    name=words[0]
    #total=int(words[1])+int(words[2])+int(words[3])+int(words[4])
    total=sum(list(map(int,words[1:])))
    print(name," ",total)
    f2.write(name+" "+str(total)+"\n")
f2.close()
```

## Regular Expressions

```python
import re
```

```python
a="mehziya"
re.match('meh',a)
```

```python
a="mehzhhiyha"
re.findall('h',a)
```

```python
a='m@ehziya'
re.search('....',a)
```

```python
a=['mehi','navya','silvaaa','dach','kavya']
for i in a:
    if re.search('^s.{4,6}$',i):
        print(i)
     
```

```python
a='mehzi32ya6'
re.search('\d+',a)
```

```python
a=['ravi','60','70','anil','80']
s=0
for i in a:
    if re.search('[0-9]+',i):
        s=int(i)+s
print(s)
```

```python
f=open('text.txt','r')
s=f.read()
m=0
for i in s:
    if re.findall('[0-9]+',i):
        m=int(i)+m
print(m)
```
